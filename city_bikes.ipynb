# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
import numpy as np
import pandas as pd
import requests

url = 'http://api.citybik.es/v2/networks'

response = requests.get(url)
data_networks = response.json()
df_networks = pd.json_normalize(data_networks['networks'])
```

```python
city_list = df_networks['location.city'].values
df_sorted = df_networks.sort_values(by = 'location.city')
canada_mask = df_sorted['location.country'] == 'CA'
df_canada = df_sorted[canada_mask]
df_canada
```
